```python
# ==============================================================================
# LIBRARY INSTALLATION AND IMPORTATION
# ==============================================================================
!pip install pandas numpy matplotlib seaborn yfinance -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi

# Configure plot style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_style("whitegrid")

print("="*70)
print(" MACRO-STRATEGIC ANALYSIS : FAST RETAILING (UNIQLO) ")
print("="*70)

# ==============================================================================
# 1.1. THE 7 FUNDAMENTAL QUESTIONS (Identity & Purpose)
# ==============================================================================
print("\n--- 1.1. The 7 questions and statements about the company ---")
questions = {
    "1. Who are we?": "Fast Retailing (Uniqlo), a global leader in casual ready-to-wear fashion.",
    "2. What do we do?": "We design, manufacture, and sell high-quality basics at affordable prices.",
    "3. Why do we do it?": "To improve everyone's daily life through the 'LifeWear' philosophy.",
    "4. Who are our customers?": "A universal public: all ages, all genders, without class distinction.",
    "5. What is our competitive advantage?": "Textile innovation (HEATTECH, AIRism) combined with the SPA model (end-to-end supply chain control).",
    "6. What are our key values?": "Quality, innovation, simplicity, longevity, and sustainability.",
    "7. Where are we going?": "To become the number one clothing brand in the world by accelerating our expansion in Europe and North America."
}

for q, a in questions.items():
    print(f"{q}\n   -> {a}")


# ==============================================================================
# 1.2. PORTER'S 5 FORCES (Web Spider Chart)
# ==============================================================================
print("\n--- 1.2. Porter's 5 Forces ---")
# Scores out of 10 (1 = Weak, 10 = Strong)
categories = ['Bargaining Power of Suppliers', 'Bargaining Power of Buyers', 'Intensity of Rivalry',
              'Threat of Substitutes', 'Threat of New Entrants']
N = len(categories)
values = [3, 6, 9, 7, 4] # Estimated values for the ready-to-wear sector
values += values[:1]     # Close the radar chart
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))
plt.xticks(angles[:-1], categories, color='grey', size=11, fontweight='bold')
ax.plot(angles, values, linewidth=2.5, linestyle='solid', color='#e60012', label="Uniqlo")
ax.fill(angles, values, color='#e60012', alpha=0.2)
ax.set_ylim(0, 10)
plt.title("Porter's 5 Forces Analysis - Uniqlo", size=15, y=1.1, fontweight='bold')
plt.show()


# ==============================================================================
# 1.3 & 1.4. SCOPE (Market Situation) & RELATIVE POSITION (Cost / Tech)
# ==============================================================================
print("\n--- 1.3 & 1.4. Scope & Relative Position of Competitors ---")

# Competitor data (Technology vs. Price)
competitors_data = {
    'Company': ['Uniqlo', 'Zara', 'H&M', 'Shein', 'Gap'],
    'Tech_Quality': [8.5, 5.5, 4.0, 2.0, 4.5],  # Score out of 10
    'Price_Level': [5.0, 6.5, 4.0, 1.5, 5.0],   # Score out of 10 (1 = very low, 10 = very expensive)
    'Market_Share': [25, 30, 20, 15, 10]        # Bubble size (Simplified Market Shares)
}
df_comp = pd.DataFrame(competitors_data)

plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df_comp, x='Tech_Quality', y='Price_Level',
    size='Market_Share', sizes=(200, 1500),
    hue='Company', palette=['#e60012', 'black', 'blue', 'orange', 'navy'],
    alpha=0.8, edgecolor='w', linewidth=2
)

# Add names next to bubbles
for i in range(len(df_comp)):
    plt.text(df_comp.Tech_Quality[i] + 0.2, df_comp.Price_Level[i],
             df_comp.Company[i], fontsize=12, fontweight='bold')

plt.title("Relative Position: Technology/Quality vs. Price Level", fontsize=15, fontweight='bold')
plt.xlabel("Technological Focus and Quality (Score /10)", fontsize=12)
plt.ylabel("Price Level (Score /10)", fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Companies")
plt.tight_layout()
plt.show()


# ==============================================================================
# 1.5 & 1.6. SWOT ANALYSIS, OPPORTUNITIES AND RISKS
# ==============================================================================
print("\n--- 1.5. SWOT Analysis ---")
swot = {
    "Strengths": ["Advanced textile R&D (HEATTECH, AIRism)", "Optimized supply chain (SPA Model)", "Strong presence in Asia"],
    "Weaknesses": ["Dependence on the Asian market", "Lower brand awareness in Europe/US compared to Zara"],
    "Opportunities": ["Rise of sustainable fashion (recycling)", "E-commerce growth", "Massive expansion in North America"],
    "Threats": ["Ultra-fast fashion (Shein, Temu)", "Geopolitical instability on the supply chain", "Rising raw material costs"]
}

for category, items in swot.items():
    print(f"\n[{category[0]}] {category} :")
    for item in items:
        print(f"  - {item}")

print("\n--- 1.6. Major Opportunities & Risks ---")
print("=> STRATEGIC OPPORTUNITY : Invest heavily in circularity and AI-based virtual try-on to boost e-commerce and reduce returns.")
print("=> STRATEGIC RISK : Disruption of global supply chains impacting raw material distribution (cotton).")


# ==============================================================================
# 1.7. EXPECTATIONS : FINANCIAL FORECASTS (Inspired by the Porsche model)
# ==============================================================================
print("\n--- 1.7. Expectations (Income Statement & Revenue Projections) ---")

# 1. Historical Data (Revenue in Billion Yen)
historical_revenue = {
    2020: 2008.0,
    2021: 2132.0,
    2022: 2301.0,
    2023: 2766.0,
    2024: 3103.0
}
df_hist = pd.DataFrame(list(historical_revenue.items()), columns=['Year', 'Revenue'])
df_hist['Type'] = 'Historical'

# Function to calculate Compound Annual Growth Rate (CAGR)
def calculate_cagr(start_val, end_val, periods):
    return (end_val / start_val) ** (1 / periods) - 1

# 2. Generate Projections
def generate_projections(years_to_project=5):
    last_year = 2024
    last_revenue = historical_revenue[last_year]

    # Growth hypotheses (YoY)
    growth_rates = {
        'Pessimistic': 0.03,  # Slow growth (economic crisis)
        'Neutral': 0.08,      # Stable growth (historical average)
        'Optimistic': 0.12    # Strong international expansion
    }

    projections = []
    for scenario, rate in growth_rates.items():
        current_rev = last_revenue
        for i in range(1, years_to_project + 1):
            current_rev *= (1 + rate)
            projections.append({'Year': last_year + i, 'Revenue': current_rev, 'Type': scenario})

    return pd.DataFrame(projections)

df_proj = generate_projections()

# Calculate CAGRs per scenario
cagr_data = {}
for scenario in ['Pessimistic', 'Neutral', 'Optimistic']:
    scen_df = df_proj[df_proj['Type'] == scenario]
    end_val = scen_df[scen_df['Year'] == 2029]['Revenue'].values[0]
    cagr_data[scenario] = calculate_cagr(historical_revenue[2024], end_val, 5)

# 3. Console display of results (Porsche style)
print("Fast Retailing (Uniqlo) Group Revenue Projections")
print("==========================================================")
print("\nHistorical Revenues (2020-2024) :")
for year, rev in historical_revenue.items():
    print(f"{year}: {rev:.1f} Billion JPY")

print("\nFinancial Analysis Summary (CAGR 2024-2029) :")
for scenario, cagr in cagr_data.items():
    print(f"Scenario {scenario} : {cagr*100:.2f}% annual growth")

# 4. Graphical Visualization
plt.figure(figsize=(12, 7))

# Historical
plt.plot(df_hist['Year'], df_hist['Revenue'], marker='o', linewidth=3, color='black', label='Historical')

# Scenarios
colors = {'Pessimistic': '#d62728', 'Neutral': '#1f77b4', 'Optimistic': '#2ca02c'}
for scenario in ['Pessimistic', 'Neutral', 'Optimistic']:
    scen_data = df_proj[df_proj['Type'] == scenario]
    # Connect the last historical point to the first projected point
    x_vals = [2024] + list(scen_data['Year'])
    y_vals = [historical_revenue[2024]] + list(scen_data['Revenue'])

    plt.plot(x_vals, y_vals, marker='s', linestyle='--', linewidth=2,
             color=colors[scenario], label=f"{scenario} (CAGR: {cagr_data[scenario]*100:.1f}%)")

plt.title("Uniqlo Revenue Projections (2020-2029)", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Revenue (Billion JPY)", fontsize=14)
plt.xticks(list(df_hist['Year']) + list(df_proj['Year'].unique()))
plt.legend(fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```
